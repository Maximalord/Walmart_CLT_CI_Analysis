```python
print("\n" + "=" * 70)
print("PART 3: CONFIDENCE INTERVAL CALCULATIONS")
print("=" * 70)
# Cell: Confidence interval function
def conf_interval(series, alpha=0.05):
    arr = series.dropna()
    n = len(arr)
    if n == 0:
        return None

    mean = arr.mean()
    se = arr.std(ddof=1) / sqrt(n)  # Standard error

    # Use z for large samples, t for small samples
    if n >= 30:
        critical_val = stats.norm.ppf(1 - alpha/2)
        method = "z"
    else:
        critical_val = stats.t.ppf(1 - alpha/2, df=n-1)
        method = "t"

    margin = critical_val * se
    return mean, mean - margin, mean + margin, n, method

print("Gender CIs (95%):")
for gender in df["Gender"].cat.categories:
    result = conf_interval(df.loc[df["Gender"] == gender, "Purchase"])
    if result:
        mean, lo, hi, n, method = result
        print(f"{gender}: n={n}, mean={mean:.2f}, CI=({lo:.2f}, {hi:.2f}), method={method}")
```

```python
print("\n" + "=" * 70)
print("PART 4: CENTRAL LIMIT THEOREM SIMULATION")
print("=" * 70)
# Cell: CLT simulation - sampling distribution of the mean
def sample_means(data, sample_size= 50, simulations = 2000):
    """Generate sample means using bootstrap sampling"""
    means = []
    for _ in range(simulations):
      sample = np.random.choice(data, size= sample_size, replace=True)
      means.append(sample.mean())

    return np.array(means)

# CLT Demonstration
# Get purchase data for first and last gender categories
male_data = df.loc[df["Gender"] == "M", "Purchase"]
female_data = df.loc[df["Gender"] == "F", "Purchase"]

sample_sizes = [10, 30, 50, 200]

for size in sample_sizes:
# Generate sampling distributions
  male_means = sample_means(male_data, sample_size=size, simulations=1500)
  female_means = sample_means(female_data, sample_size=size, simulations=1500)

# Create side-by-side plots
  fig, (ax1, ax2) = plt.subplots(1,2, figsize= (12,4))
  sns.histplot(male_means, bins=30, kde=True, ax=ax1)
  ax1.set_title(f"Male (n={size})")

  sns.histplot(female_means, bins=30, kde=True, ax=ax2)
  ax2.set_title(f"Female (n={size})")

  plt.tight_layout()
  plt.show()
```

```python
print("\n" + "=" * 70)
print("PART 5: MULTI-LEVEL CONFIDENCE INTERVALS (90%, 95%, 99%)")
print("=" * 70)
def group_confidence_intervals_multilevel(df, group_column):
    """Calculate confidence intervals for purchase amounts by group at 90%, 95%, 99% levels"""
    # Get unique groups from the column
    groups = df[group_column].unique()
    results = []

    for group in groups:
        # Get purchase data for this group
        group_data = df.loc[df[group_column] == group, "Purchase"]

        # Calculate confidence intervals for different levels
        ci_90 = conf_interval(group_data, alpha=0.10)  # 90% CI
        ci_95 = conf_interval(group_data, alpha=0.05)  # 95% CI
        ci_99 = conf_interval(group_data, alpha=0.01)  # 99% CI

        results.append({
            "group": group,
            "n": len(group_data),
            "mean": group_data.mean(),
            "CI_90_lower": ci_90[1],
            "CI_90_upper": ci_90[2],
            "CI_95_lower": ci_95[1],
            "CI_95_upper": ci_95[2],
            "CI_99_lower": ci_99[1],
            "CI_99_upper": ci_99[2]
        })

    return pd.DataFrame(results)

# Generate CI tables for all demographic groups with multiple confidence levels
print("Gender - Confidence Intervals (90%, 95%, 99%):")
gender_ci_df = group_confidence_intervals_multilevel(df, "Gender")
display(gender_ci_df)

# Print formatted results for Gender (M and F)
print("\n**Confidence Intervals (90%,95%,99%):**")
for _, row in gender_ci_df.iterrows():
    group = row['group']
    print(f"- 90% CI â€” {group}: ({row['CI_90_lower']:.2f}, {row['CI_90_upper']:.2f})")
    print(f"- 95% CI â€” {group}: ({row['CI_95_lower']:.2f}, {row['CI_95_upper']:.2f})")
    print(f"- 99% CI â€” {group}: ({row['CI_99_lower']:.2f}, {row['CI_99_upper']:.2f})")
```

**Confidence Intervals (90%,95%,99%):**
- 90% CI â€” M: (9424.51, 9450.54); F: (8713.29, 8755.84)
- 95% CI â€” M: (9422.02, 9453.03); F: (8709.21, 8759.92)
- 99% CI â€” M: (9417.15, 9457.91); F: (8701.24, 8767.89)


```python
print("\n" + "=" * 70)
print("PART 6: STATISTICAL SIGNIFICANCE TESTING  & TWO SAMPLE T TEST (male vs female)")
print("=" * 70)

def intervals_overlap(a_lo, a_hi, b_lo, b_hi):
    """Check if two confidence intervals overlap"""
    return not (a_hi < b_lo or b_hi < a_lo)

# Check CI overlap using 95% CI
m_row = gender_ci_df[gender_ci_df['group'] == 'M'].iloc[0]
f_row = gender_ci_df[gender_ci_df['group'] == 'F'].iloc[0]

print("Confidence Interval Overlap Check (95% CI):")
print(f"Group M: mean={m_row['mean']:.2f}, 95% CI=({m_row['CI_95_lower']:.2f}, {m_row['CI_95_upper']:.2f}), n={m_row['n']}")
print(f"Group F: mean={f_row['mean']:.2f}, 95% CI=({f_row['CI_95_lower']:.2f}, {f_row['CI_95_upper']:.2f}), n={f_row['n']}")
print("Do 95% CIs overlap?", intervals_overlap(m_row['CI_95_lower'], m_row['CI_95_upper'],
                                              f_row['CI_95_lower'], f_row['CI_95_upper']))

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(male_data, female_data, equal_var=False)
print(f"\nTwo-sample t-test (unequal variances):")
print(f"t-statistic = {t_stat:.3f}, p-value = {p_value:.4f}")
```

**Business implication:** CIs do not overlap â€” evidence of a difference in mean spend by gender; consider targeted offers.
